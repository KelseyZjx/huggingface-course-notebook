```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
context = """
🤗 Transformers is backed by the three most popular deep learning libraries — Jax, PyTorch, and TensorFlow — with a seamless integration
between them. It's straightforward to train your models with one before loading them for inference with the other.
"""
question = "Which deep learning libraries back 🤗 Transformers?"
question_answerer(question=question, context=context)
```

```python
#让我们处理一个非常长的文本试一试
long_context = """
🤗 Transformers: State of the Art NLP

🤗 Transformers provides thousands of pretrained models to perform tasks on texts such as classification, information extraction,
question answering, summarization, translation, text generation and more in over 100 languages.
Its aim is to make cutting-edge NLP easier to use for everyone.

🤗 Transformers provides APIs to quickly download and use those pretrained models on a given text, fine-tune them on your own datasets and
then share them with the community on our model hub. At the same time, each python module defining an architecture is fully standalone and
can be modified to enable quick research experiments.

Why should I use transformers?

1. Easy-to-use state-of-the-art models:
  - High performance on NLU and NLG tasks.
  - Low barrier to entry for educators and practitioners.
  - Few user-facing abstractions with just three classes to learn.
  - A unified API for using all our pretrained models.
  - Lower compute costs, smaller carbon footprint:

2. Researchers can share trained models instead of always retraining.
  - Practitioners can reduce compute time and production costs.
  - Dozens of architectures with over 10,000 pretrained models, some in more than 100 languages.

3. Choose the right framework for every part of a model's lifetime:
  - Train state-of-the-art models in 3 lines of code.
  - Move a single model between TF2.0/PyTorch frameworks at will.
  - Seamlessly pick the right framework for training, evaluation and production.

4. Easily customize a model or an example to your needs:
  - We provide examples for each architecture to reproduce the results published by its original authors.
  - Model internals are exposed as consistently as possible.
  - Model files can be used independently of the library for quick experiments.

🤗 Transformers is backed by the three most popular deep learning libraries — Jax, PyTorch and TensorFlow — with a seamless integration
between them. It's straightforward to train your models with one before loading them for inference with the other.
"""
question_answerer(question=question, context=long_context)
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering

model_checkpoint = "distilbert-base-cased-distilled-squad"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForQuestionAnswering.from_pretrained(model_checkpoint)

inputs = tokenizer(question, context, return_tensors="pt")
outputs = model(**inputs)
```

```python
#与其他任务不同，问答模型返回两个张量而不是一个（一个表示预测答案开始的标记索引，另一个表示答案结束处的标记索引）
start_logits = outputs.start_logits
end_logits = outputs.end_logits
print(start_logits.shape, end_logits.shape)
```

为了将这些Logits转换为概率，我们将应用softmax函数。

在此之前，为了确保屏蔽我们不想处理的、不属于上下文的索引[CLS]、[SEP]，我们需要屏蔽[SEP]标记。但是我们将保留[CLS]标记，因为某些模型使用它指示答案不在上下文中。

我们只需要把将要屏蔽的logits替换为一个超大的负数，在这里设置为-10000，after softmax,该值即变为了0

```python
import torch

sequence_ids = inputs.sequence_ids()
#mask everything apart from the tokens of the context
mask = [i != 1 for i in sequence_ids]
#unmask the [CLS] token
mask[0] = False
mask = torch.tensor(mask)[None]

start_logits[mask] = -10000
end_logits[mask] = -10000
```

```python
print(mask)
```

```python
#我们已经适当的屏蔽了我们不想预测的位置的分数，现在可以开始使用softmax函数了
start_probabilities = torch.nn.functional.softmax(start_logits, dim=-1)[0]
end_probabilities = torch.nn.functional.softmax(end_logits, dim=-1)[0]
```

在此，我们可以取开始和结束概率的argmax，但是我们有可能取到一个开始索引比结束索引大的结果。所以我们在此采取几个预防措施。

我们将会计算所有的start_index<=end_index的start_index和end_index概率，然后取出概率最高的(start_index,end_index)概率。

假设事件“The answer starts at start_index” and “The answer ends at end_index”是相互独立的，answer从start_index开始并在end_index结束的概率为：

start_probabilities[start_index] × end_probabilities[end_index]

所以，想要计算得分，我们只需要计算所有start_index <= end_index情况下的start_probabilities[start_index] × end_probabilities[end_index]即可。

```python
#compute all possible products
scores = start_probabilities[:, None] * end_probabilities[None, :]
```

```python
#torch.triu()返回2D张量的上三角部分
scores = torch.triu(scores)
```

```python
#找最大值
max_index = scores.argmax().item()
start_index = max_index // scores.shape[1]
end_index = max_index % scores.shape[1]
print(scores[start_index, end_index])
```

我们有用token表示的答案的start_index和end_index，所以现在我们只需要转换为上下文中的字符索引。

这就是offsets非常有用的地方，我们可以抓取使用它们。

```python
inputs_with_offsets = tokenizer(question, context, return_offsets_mapping=True)
offsets = inputs_with_offsets["offset_mapping"]

start_char, _ = offsets[start_index]
_, end_char = offsets[end_index]
answer = context[start_char:end_char]
```

```python
result = {
    "answer": answer,
    "start": start_char,
    "end": end_char,
    "score": scores[start_index, end_index],
}

print(result)
```

```python
#handling long context
inputs = tokenizer(question, long_context)
print(len(inputs["input_ids"]))
```

如果我们尝试分词之前的长文本示例，我们将会得到一个比question-answering pipeline的最大限制长度（384）更长的文本。

因此我们需要进行截断。我们有很多方法可以做到这一点，但是我们不想截断问题，只想截断上下文。

由于上下文是第二个句子，我们可以使用"only_second"截断策略。

但这么做的毛病在于：问题的答案可能不在被截断的上下文中。例如下面这个例子我们选择了一个问题，其答案位于上下文末尾，而当我们截断它时，答案不存在。

```python
inputs = tokenizer(question, long_context, max_length=384, truncation="only_second")
print(tokenizer.decode(inputs["input_ids"]))
```

这意味着模型将很难选择正确的答案。

为了解决这个问题，answering pipeline允许我们将上下文分割成更小的块，并指定最大长度。为了确保我们不会在错误的地方分割上下文以找到答案，它还包括块之间的一些重叠。

我们可以通过添加return_overflowing_to_keys=True让tokenizer（快速或慢速）来完成此操作，并且可以通过定义stride指定我们想要的重叠。

```python
sentence = "This sentence is not too long but we are going to split it anyway."
inputs = tokenizer(
    sentence, truncation=True, return_overflowing_tokens=True, max_length=6, stride=2
)

for ids in inputs["input_ids"]:
    print(tokenizer.decode(ids))
```

```python
#overflow_to_sample_mapping告诉我们结果来自于哪个句子。
#我们得到7个结果都来自同一个句子

print(inputs["overflow_to_sample_mapping"])
```

```python
#当我们token多个句子时，它的作用就体现出来了
sentences = [
    "This sentence is not too long but we are going to split it anyway.",
    "This sentence is shorter but will still get split.",
]
inputs = tokenizer(
    sentences, truncation=True, return_overflowing_tokens=True, max_length=6, stride=2
)

print(inputs["overflow_to_sample_mapping"])
```

```python
inputs = tokenizer(
    question,
    long_context,
    stride=128,
    max_length=384,
    padding="longest",
    truncation="only_second",
    return_overflowing_tokens=True,
    return_offsets_mapping=True,
)
```

```python
#弹出overflow_to_sample_mapping和offset_mapping这两个我们不需要使用的参数
_ = inputs.pop("overflow_to_sample_mapping")
offsets = inputs.pop("offset_mapping")

inputs = inputs.convert_to_tensors("pt")
print(inputs["input_ids"].shape)
```

```python
#long context被分成了2部分，意味着会有两组start and end logits
outputs = model(**inputs)

start_logits = outputs.start_logits
end_logits = outputs.end_logits
print(start_logits.shape, end_logits.shape)
```

```python
sequence_ids = inputs.sequence_ids()
# Mask everything apart from the tokens of the context
mask = [i != 1 for i in sequence_ids]
# Unmask the [CLS] token
mask[0] = False
# Mask all the [PAD] tokens
mask = torch.logical_or(torch.tensor(mask)[None], (inputs["attention_mask"] == 0))

start_logits[mask] = -10000
end_logits[mask] = -10000
```

```python
start_probabilities = torch.nn.functional.softmax(start_logits, dim=-1)
end_probabilities = torch.nn.functional.softmax(end_logits, dim=-1)
```

```python
#这里我们做的和之前做过的类似，但是对两个部分重复：取两个部分的最高得分
candidates = []
for start_probs, end_probs in zip(start_probabilities, end_probabilities):
    scores = start_probs[:, None] * end_probs[None, :]
    idx = torch.triu(scores).argmax().item()

    start_idx = idx // scores.shape[1]
    end_idx = idx % scores.shape[1]
    score = scores[start_idx, end_idx].item()
    candidates.append((start_idx, end_idx, score))

print(candidates)
```

```python
for candidate, offset in zip(candidates, offsets):
    start_token, end_token, score = candidate
    start_char, _ = offset[start_token]
    _, end_char = offset[end_token]
    answer = long_context[start_char:end_char]
    result = {"answer": answer, "start": start_char, "end": end_char, "score": score}
    print(result)
```
